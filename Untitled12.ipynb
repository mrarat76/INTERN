```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Veriyi tarihe göre sıralayalım
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# Tarlaları ayrı ayrı zaman serisi modelleriyle işleyebiliriz. Öncelikle bir tarlayı ele alalım:
tarla_1 = df[df['Polygon Name'] == 'mısır_1']

# NDVI, NDWI ve LAI verilerinin zaman serisini oluşturalım
# NDVI, NDWI ve LAI verilerinin zaman serisini oluşturalım
ndvi_series = tarla_1.set_index('Date')['Mean NDVI (Original Polygon)']
ndwi_series = tarla_1.set_index('Date')['Mean NDWI (Original Polygon)']
lai_series = tarla_1.set_index('Date')['Mean LAI (Original Polygon)']

# Zaman serisi modelini eğitmek için NDVI verisini kullanacağız (SARIMA ile):
# Burada SARIMA(p,d,q)(P,D,Q,s) modelinin parametrelerini optimize edebilirsiniz.
sarima_model = SARIMAX(ndvi_series, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_result = sarima_model.fit()

# Modelin sonuçlarını inceleyelim
print(sarima_result.summary())

# Modelden tahmin yapalım (geçmişe ve geleceğe yönelik):
ndvi_forecast = sarima_result.get_forecast(steps=12)
forecast_index = pd.date_range(start=ndvi_series.index[-1], periods=12, freq='M')
forecast_mean = ndvi_forecast.predicted_mean
forecast_ci = ndvi_forecast.conf_int()

# Tahmin sonuçlarını görselleştirelim
plt.figure(figsize=(10,6))
plt.plot(ndvi_series.index, ndvi_series, label='Gerçek NDVI')
plt.plot(forecast_index, forecast_mean, label='Tahmin Edilen NDVI')
plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='r', alpha=0.3)
plt.title('NDVI Tahminleri')
plt.legend()
plt.show()

# Hasat ile ilişkilendirme (örneğin en son 3 ayın NDVI ortalamasını kullanabiliriz)
average_ndvi_last_3_months = ndvi_series[-3:].mean()

# Burada basit bir ilişkilendirme yapabiliriz. Gerçek hasat verisiyle tahmini karşılaştırabiliriz.
# Örnek hasat verisi (mısır_1 tarlası için 500 birim hasat)
actual_harvest = 26905

# Önerilen tahmin (örneğin: NDVI ortalamasına dayalı basit bir model)
predicted_harvest = average_ndvi_last_3_months * 1000  # Bu katsayı sizin modelinize göre ayarlanabilir

print(f"Gerçek Hasat: {actual_harvest}")
print(f"Tahmin Edilen Hasat: {predicted_harvest}")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Zaman serisini hazırlayalım
ndvi_series = df[df['Polygon Name'] == 'mısır_1'].set_index('Date')['Mean NDVI (Original Polygon)']

# Veriyi ölçeklendirelim (LSTM modelleri genellikle normalleştirilmiş verilerle daha iyi çalışır)
scaler = MinMaxScaler(feature_range=(0, 1))
ndvi_scaled = scaler.fit_transform(ndvi_series.values.reshape(-1, 1))

# Zaman serisi için veri pencereleri oluşturma (örn. son 30 günün verisiyle tahmin yapalım)
def create_dataset(dataset, time_step=30):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)

time_step = 30  # Son 30 günün verisi ile tahmin yapılacak
X, y = create_dataset(ndvi_scaled, time_step)

# Veriyi yeniden şekillendirelim [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Eğitim ve test setlerine ayıralım (80% eğitim, 20% test)
train_size = int(len(X) * 0.8)
test_size = len(X) - train_size
X_train, X_test = X[0:train_size], X[train_size:len(X)]
y_train, y_test = y[0:train_size], y[train_size:len(y)]

# LSTM modelini oluşturalım
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Modeli derleyelim
model.compile(optimizer='adam', loss='mean_squared_error')

# Modeli eğitelim
model.fit(X_train, y_train, batch_size=64, epochs=100)

# Modelin tahminini yapalım
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Veriyi orijinal ölçeğe geri çevirelim
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train_orig = scaler.inverse_transform([y_train])
y_test_orig = scaler.inverse_transform([y_test])

# Tahmin sonuçlarını görselleştirelim
plt.figure(figsize=(10,6))
plt.plot(ndvi_series.index[-len(y_test):], y_test_orig[0], label='Gerçek NDVI')
plt.plot(ndvi_series.index[-len(test_predict):], test_predict, label='Tahmin Edilen NDVI')
plt.title('Gerçek ve Tahmin Edilen NDVI (LSTM)')
plt.xlabel('Tarih')
plt.ylabel('NDVI')
plt.legend()
plt.show()
```
